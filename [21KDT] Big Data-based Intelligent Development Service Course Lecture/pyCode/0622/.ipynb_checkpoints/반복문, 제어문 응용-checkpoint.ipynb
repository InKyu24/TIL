```python
from PIL import Image
import pandas as pd
flieName = ["pic1", "pic2", "pic3"]
prnData = []

for fn in flieName:
    fName = fn+'.jpg'
    tmp = Image.open("./picture/"+ fName)
    fm = tmp.format
    print(tmp)
    fw = tmp.width
    fh = tmp.height
    prnData.append([fName, fm, fw, fh])
    
    tmpRes = tmp.resize((300,300))
    tmpRes.save('./data/'+fn+"_300.jpg")
print("--END--")
print(prnData)

df = pd.DataFrame(prnData, columns=["파일명", "형식", "너비", "높이"])
df.to_csv('./파일정보.csv', encoding="cp949")
display(df)
```

```python
#picture 폴더 파일명 전체 가져오기
# 그림 파일만 체크해서 읽기
# 각 파일별 파일명, 사이즈를 저장하기 (각 파일별 속성확인 flieName format w h)
# 각 파일별 사이즈 300 300 변경 후 data 폴더에 저장
# 파일명_300.jpg
```

```python
from PIL import Image
import pandas as pd

fileList = os.listdir('./picture')

fileInfo = []

for file in fileList:
    fn = file.split(".") # 파일명과 확장자 분리
    name = fn[0] # 파일명
    form = fn[1] # 확장자
    if len(fn) > 1 :# 확장자가 있는 파일에 대해서
        if form == "jpg": # 확장자가 jpg인 경우에
            tmp = Image.open('./picture/'+file)# 이미지 열기
            fileInfo.append([file, tmp.format, tmp.width, tmp.height]) # 해당 이미지 파일정보를 리스트에 추가
            
            tmpRes = tmp.resize((300, 300)) # 해당 이미지 사이즈 조절하고,
            tmpRes.save('./data/'+name+'_300.'+form) # 조정된 이미지 저장
            
df = pd.DataFrame(fileInfo, columns=["파일명", "형식", "너비", "높이"]) # 파일정보 리스트에 열 추가
df.to_csv('./fileInfo.csv', encoding="cp949") # 파일로 저장
display(df) # 화면 출력

```
